8.5.1: the pixels are transformed to the average value of its neighbors
8.5.2: the -1 represents the parameter for the anchor:
    "anchor  anchor of the kernel that indicates the relative position of a filtered point within the kernel; the anchor should lie within the kernel; default value (-1,-1) means that the anchor is at the kernel center." (openCV documentation)
8.5.3: I found a stack overflow answer that comes from OpenCVs documentation, the first number is the bitdepth, the letter following can either be U S or F, representing unsigned int, signed int, or float type. so 8u is 8 bit depth, and unsigned int, 16S is signed int, and 16 bit depth. cv_64F is floating point type with 64bit color. The bit depth describes the number of bits used in colors to acheive that color, the more bits the larger range of colors.

```python
# 8.8
# load libs
import cv2
import numpy as np
from matplotlib import pyplot as plt

# load image
image_bgr = cv2.imread("images/plane.jpg")
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

# define range of blue values
lower_blue = np.array([50,100,50])
upper_blue = np.array([130,255,255])

# create mask
mask = cv2.inRange(image_hsv, lower_blue, upper_blue)

# mask image
image_bgr_masked = cv2.bitwise_and(image_bgr, image_bgr, mask=mask)

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image_bgr_masked, cv2.COLOR_BGR2RGB)

# show image
plt.imshow(image_rgb), plt.axis("off")
plt.show()

# show image gray
plt.imshow(mask, cmap='gray'), plt.axis("off")
plt.show()
```

8.8.1: hue, saturation, value
8.8.2: it is converting them to the desired output, and also flipping the binary values if 0 and 1

```python
# 8.9
# load libs
import cv2
import numpy as np
from matplotlib import pyplot as plt

# laod img grayscale
image_grey = cv2.imread("images/plane.jpg", cv2.IMREAD_GRAYSCALE)

# Apply adaptive thresholding
max_output_value = 255
neighborhood_size = 99
subtract_from_mean = 10
image_binarized = cv2.adaptiveThreshold(image_grey, max_output_value, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, neighborhood_size, subtract_from_mean)

# show image
plt.imshow(image_binarized, cmap="gray"), plt.axis("off")
plt.show()
```

```python
# apply cv2.Adaptive thresh mean
image_mean_threshold = cv2.adaptiveThreshold(image_grey, max_output_value, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, neighborhood_size, subtract_from_mean)

# show image
plt.imshow(image_mean_threshold, cmap="gray"), plt.axis("off")
plt.show()
```

8.9.1: basically, if a pixels value is greater than a certain amount than it becomes white, and if it is less than it becomes black which then helps to isolate objects in the image.
8.9.2:we can isolate the text much better that way and eliminate the background

```python
# 8.10
# load libs
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_bgr = cv2.imread("images/plane.jpg")
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# rectangle vals
rectangle = (0, 56, 256, 150)

# create mask
mask = np.zeros(image_rgb.shape[:2], np.uint8)

# create temp arrays
bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)

# run grabCut
cv2.grabCut(image_rgb, mask, rectangle, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

mask_2 = np.where((mask==2) | (mask == 0), 0, 1).astype('uint8')

image_rgb_nobg = image_rgb * mask_2[:, :, np.newaxis]

plt.imshow(image_rgb_nobg), plt.axis("off")
plt.show()

# show mask
plt.imshow(mask, cmap='gray'), plt.axis("off")
plt.show()

# show other mask
plt.imshow(mask_2, cmap='gray'), plt.axis("off")
plt.show()
```

```python
# 8.11
# load libs
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_gray = cv2.imread("images/plane.jpg", cv2.IMREAD_GRAYSCALE)

# calc median intensity
median_intensity = np.median(image_gray)

lower_threshold = int(max(0, (1.0-.33)* median_intensity))
upper_threshold = int(min(255, (1.0 + .33) * median_intensity))

image_canny = cv2.Canny(image_gray, lower_threshold, upper_threshold)

# show image
plt.imshow(image_canny, cmap="gray"), plt.axis("off")
plt.show()
```

```python
# 8.12
# load libs
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_bgr = cv2.imread("images/plane.jpg")
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
image_gray = np.float32(image_gray)

block_size = 2
apeture = 29
free_parameter = 0.04

detector_responses = cv2.cornerHarris(image_gray, block_size, apeture, free_parameter)

detector_responses = cv2.dilate(detector_responses, None)

threshold = 0.02
image_bgr[detector_responses > threshold * detector_responses.max()] = [255,255,255]

image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(image_gray, cmap="gray"), plt.axis("off")

plt.imshow(detector_responses, cmap="gray"), plt.axis("off")
plt.show()

image_bgr = cv2.imread("images/plane.jpg")
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

corners_to_detect = 10
minimum_quality_score = 0.05
minimum_distance = 25

corners = cv2.goodFeaturesToTrack(image_gray, corners_to_detect, minimum_quality_score, minimum_distance)

corners = np.float32(corners)

for corner in corners:
    x, y = corner[0]
    cv2.circle(image_bgr, (x,y), 10, (255,255,255), -1)
    
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(image_rgb, cmap='gray'), plt.axis("off")
plt.show()
```
